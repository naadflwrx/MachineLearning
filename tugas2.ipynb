```python
# memanggil dataset lewat gdrive
path = "/content/gdrive/MyDrive/Praktikum_ML/pertemuan-2"

# membaca file csv menggunakan pandas
import pandas as pd
df = pd.read_csv(path + "/data/day.csv")
df
```

```python
# import library untuk membagi dataset
from sklearn.model_selection import train_test_split

# menghitung jumlah data
total_size = len(df)
train_size = int(total_size * 0.8)
test_size = int(total_size * 0.2)

# membagi dataset menjadi training dan testing
train_df, test_df = train_test_split(df, test_size=test_size, random_state=42)

# membagi data training menjadi train final dan validation (10% dari training)
val_size = int(len(train_df) * 0.1)  # 10% dari training untuk validation
train_df, val_df = train_test_split(train_df, test_size=val_size, random_state=42)

# menampilkan jumlah data setiap set
print("Jumlah data Training\t:", len(train_df))
print("Jumlah data Validation\t:", len(val_df))
print("Jumlah data Testing\t:", len(test_df))
```

```python
# menampilkan 5 baris teratas setiap set
print("\n5 baris data teratas Data Training:")
display(train_df.head())

print("\n5 baris data teratas Data Validation:")
display(val_df.head())

print("\n5 baris data teratas Data Testing:")
display(test_df.head())
```

```python
import matplotlib.pyplot as plt

# Data jumlah sampel
sets = ['Training', 'Validation', 'Testing']
counts = [len(train_df), len(val_df), len(test_df)]
total = sum(counts)

# Hitung persentase
percentages = [count / total * 100 for count in counts]

# Warna vintage cerah
colors = ['#E1BFB7', '#F6E7DC', '#A3B5B0']

# Pie chart
plt.figure(figsize=(6,6))
plt.pie(percentages, labels=sets, autopct='%1.1f%%', colors=colors,
        startangle=140, explode=(0.05, 0.05, 0.05), shadow=True)
plt.title('Persentase Dataset', fontsize=14, fontweight='bold')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data jumlah sampel
sets = ['Training', 'Validation', 'Testing']
counts = [len(train_df), len(val_df), len(test_df)]

# Warna vintage cerah, sama seperti pie chart
colors = ['#E1BFB7', '#F6E7DC', '#A3B5B0']

# Bar chart
plt.figure(figsize=(6,5))
bars = plt.bar(sets, counts, color=colors, edgecolor='black')

# Menambahkan label jumlah di atas batang
for bar, count in zip(bars, counts):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 1, str(count),
             ha='center', va='bottom', fontsize=12)

plt.title('Jumlah Data per Set', fontsize=14, fontweight='bold')
plt.ylabel('Jumlah Sampel')
plt.ylim(0, max(counts)*1.2)  # memberi ruang untuk label di atas batang
plt.show()
```
